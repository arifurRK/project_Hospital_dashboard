```python
class PostgresConnection(object):
    def __init__(self):
        self.connection = psycopg2.connect(database="hospital",
                                           user = "postgres",
                                           password = "system",
                                           host = "localhost",
                                           port = "5432")

    def getConnection(self):
        print("Connection to DB established!")
        return self.connection
```

```python
con = PostgresConnection().getConnection()
```

```python
con = PostgresConnection().getConnection()
insert_stmt = "INSERT INTO hospitals.patient_info(pik,APPOINT_NO,AGE_YY,GENDER,ADDRESS,DEPARTMENT_NO)VALUES (%s, %s, %s, %s, %s ,%s);"
cur = con.cursor()
psycopg2.extras.execute_batch(cur, insert_stmt, patient_info.values)
con.commit()
cur.close()
```

```python
con = PostgresConnection().getConnection()
insert_stmt = "INSERT INTO hospitals.doctor_info(DOCTOR_NO,DEPARTMENT_NAME)VALUES (%s, %s);"
cur = con.cursor()
psycopg2.extras.execute_batch(cur, insert_stmt, doctor_info.values)
con.commit()
cur.close()
```

```python
con = PostgresConnection().getConnection()
insert_stmt = "INSERT INTO hospitals.department_info(DEPARTMENT_NO,DEPARTMENT_NAME)VALUES (%s, %s);"
cur = con.cursor()
psycopg2.extras.execute_batch(cur, insert_stmt, department_info.values)
con.commit()
cur.close()
```

```python
con = PostgresConnection().getConnection()
insert_stmt = "INSERT INTO hospitals.bill_info(APPOINT_NO,DEPARTMENT_NO,DOCTOR_NO,CONSULT_FEE,DAY,MONTH,DATE,YEAR)VALUES (%s, %s, %s, %s, %s,%s, %s, %s);"
cur = con.cursor()
psycopg2.extras.execute_batch(cur, insert_stmt, bill_info.values)
con.commit()
cur.close()
```
